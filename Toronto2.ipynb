```python
import pandas as pd
```

```python
postcode_df = pd.read_csv("Geospatial_Coordinates.csv")
read_df = pd.read_csv("torontof.csv") # "torontof.csv" is where I stored the previous table
```

```python
postcode_col = postcode_df.columns.values
postcode_col[0] = read_df.columns.values[0]
```

```python
merged_df = pd.merge(read_df,postcode_df,on= 'Postcode')
```

```python
merged_df.head()
```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import folium 
```

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
borough_data = merged_df[merged_df['Borough'].str.contains("Toronto")]
borough_data.head()
```

```python
# add markers to map
for lat, lng, label in zip(borough_data['Latitude'], borough_data['Longitude'], borough_data['Borough']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
